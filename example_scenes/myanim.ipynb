```python
from manim import *
# set the maximum width for video outputs to a predefined value
config.media_width = "75%"
# embed video
config.media_embed = True
```

```python
%%manim -v WARNING --disable_caching -qm MyScene
import numpy as np
import random


class MyScene(Scene):
    def construct(self):
        
        title = Tex(r"KCCS presents")
        self.play(Write(title))
        self.wait()

        transform_title = Tex("This is our system")
        sub_title = Tex(r"Here is sub title")
        VGroup(transform_title, sub_title).arrange(DOWN).to_corner(UL)
        
        self.play(
            Transform(title, transform_title),
            LaggedStart(FadeIn(sub_title, shift=UP)),
        )
        self.wait()

        # Step 1: An image fades in.
        img = ImageMobject("G:/OneDrive/AI/NovelAI/{{{1 ghost girl}}}, {{{ghost costume}}}, {{ghost tail}}, ghost slave, ghost hair ornament, {{{aura}}}, {translucent body}, {floating}, floating skirt, flying, {{graveyard}}, skeb c s-926054877 (1).png")
        self.play(FadeIn(img))
        self.wait()

        # Draw 20 small squares randomly in the image
        squares = []
        for i in range(20):
            x = random.uniform(-img.width / 2, img.width / 2)
            y = random.uniform(-img.height / 2, img.height / 2)
            square = Square(side_length=0.2, color=RED, fill_opacity=0).move_to([x, y, 0])
            squares.append(square)
        square_g = VGroup(*squares)
        self.play(Create(square_g))
        self.wait()

        # Select first 5 samples and zoomed
        zoomed_squares, selected_squares, removed_squares = [], [], []
        for i, square in enumerate(squares):
            if i < 5:
                selected_squares.append(square)
                zoomed_square = square.copy().scale(4).to_edge(RIGHT)
                zoomed_squares.append(zoomed_square)
                self.play(square.animate.scale(2).set_color(YELLOW), run_time=0.5)
            else:
                removed_squares.append(square)
        selected_squares = VGroup(*selected_squares)
        removed_squares = VGroup(*removed_squares)
        zoomed_squares_group = VGroup(*zoomed_squares).arrange_submobjects(RIGHT, buff=0.4)
        self.play(
            FadeOut(img),
            Transform(selected_squares, zoomed_squares_group),
            FadeOut(removed_squares),
        )
        self.wait()
```

```python
%%manim -v WARNING --disable_caching -qm CutoutExample

class CutoutExample(Scene):
    def construct(self):
        title = Tex(r"KCCS presents")
        self.play(Write(title))
        self.wait()

        transform_title = Tex("That was a our system")
        transform_title.to_corner(UL)
        
        self.play(
            Transform(title, transform_title),
        )
        self.wait()

        

        s1 = Square().scale(2.5)
        s2 = Triangle().shift(DOWN + RIGHT).scale(0.5)
        s3 = Square().shift(UP + RIGHT).scale(0.5)
        s4 = RegularPolygon(5).shift(DOWN + LEFT).scale(0.5)
        s5 = RegularPolygon(6).shift(UP + LEFT).scale(0.5)
        c = Cutout(s1, s2, s3, s4, s5, fill_opacity=0.5, color=BLUE, stroke_color=RED)
        self.play(Write(c), run_time=3)
        self.wait()
```
